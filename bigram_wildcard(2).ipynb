```python
def rempun(s):
  s=str(s)
  punc = '''!()-[]{};:'"\, <>./?@#$%^&*_~'''
  for i in range(len(s)):
    if s[i] in punc:
      s=s.replace(s[i]," ")
  s=s.lower()
  return s
d1=rempun(d1)
d2=rempun(d2)
d3=rempun(d3)
print(d1,"\n",d2,"\n",d3)
```

```python
def split(d):
  l1=d.split()
  res=[]
  for i in l1:
    if len(i)==1:
      res.append(i)
    else:
      res.append('$'+i[0])
      for j in range(0,len(i)-1):
        res.append(i[j:j+2])
      res.append(i[len(i)-1]+'$')
  return res
b1=split(d1)
b2=split(d2)
b3=split(d3)
print(b1,"\n",b2,"\n",b3)
```

```python
def intersect(p1,p2):
  res=[]
  for i in p1:
    if i in p2:
      res.append(i)
  return res
```
